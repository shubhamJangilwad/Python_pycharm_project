```python
def print_num(n):
    if n > 5:
        return n
    print(n)
    print_num(n+1)

print_num(1)
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n*factorial(n-1)
print(factorial(8))
```

```python
def sum_num(n):
    if n == 0:
        return 0
    return n + sum_num(n-1)
print(sum_num(20))
```

```python
def sum_digits(n):
    if n == 0:
        return 0

    return n % 10 + sum_digits(n // 10)


print(sum_digits(1234))
```

```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1

    return fib(n-1) + fib(n-2)


print(fib(6))
```
